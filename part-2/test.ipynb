```python
import os
# from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.inception_v3 import preprocess_input, decode_predictions
import numpy as np
import json
folder_path="test/" #provide the test path
folder_path+="anonymous/"
def load_model():
    # Specify the folder containing images
    model = tf.keras.models.load_model("dlTrained_model.h5")
    return model

def getprediction(model,folder_path,target_size,class_names):
    d={}
    # Iterate over all images in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith('.jpg') or filename.endswith('.png'):  # Adjust the file extensions as needed
            # Load and resize the image
            img_path = os.path.join(folder_path, filename)
            img = image.load_img(img_path, target_size=target_size)
            img_array = image.img_to_array(img)
            img_array = np.expand_dims(img_array, axis=0)
            img_array = preprocess_input(img_array)

            # Make predictions
            predictions = model.predict(img_array)

            # Get the predicted class index
            predicted_class_index = np.argmax(predictions)
            
            # Get the corresponding label
            predicted_label = class_names[predicted_class_index]
            d[filename]=predicted_label
    return d

def saveJsonfile(d):      
    output_file_path = 'output.json'
    with open(output_file_path, 'w') as json_file:
        json.dump(d, json_file, indent=4)
        
if __name__=='__main__':
    target_size = (180, 180)
    class_names=['creamy_paste', 'diced', 'floured', 'grated', 'juiced', 'jullienne', 'mixed', 'other', 'peeled', 'sliced', 'whole']
    model=load_model()
    output=getprediction(model,folder_path,target_size,class_names)
    saveJsonfile(output)
    
    
```
